```python
# fibonacci

a=0
b=1

conta=0

print (a)
print (b)

while conta<10:
    c=a+b
    print (c)
    
    a=b
    b=c
    
    conta+=1
```

```python
lst_lst=[]

for i in range(0, 9, 3):
    lst_lst.append([i+1, i+2, i+3])
    
print (lst_lst)
```
